```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Data Sests
### 1.Tips (Sample Dataset)
### 2. COVID-19( (Real-world Dataset)


```
tips = sns.load_dataset("tips")
tips.head()
```

```
tips.shape  #get the size of the database
```

Count plot of day

```
sns.countplot(tips.day)
plt.show()
```

```
sns.countplot(tips.day, palette="cubehelix")
plt.show()
```

```
sns.countplot(tips.day, palette=["b","r","sandybrown", "gray"])
plt.show()
```

```
# visualize the smokers in the dataset
```

```
sns.countplot(tips.smoker)
plt.show()
```

barplot with tip by day

```
sns.barplot(x="day", y="tip", data=tips)
plt.show()
```

```
sns.set(rc={'figure.figsize': (12,8)}) #change the figure size
sns.barplot(x="day", y="tip", data=tips, ci=None)

plt.show()
```

```
sns.set_style('darkgrid')
sns.set_palette("RdBu")
sns.set_context("poster")

sns.barplot(x="day", y="tip", data=tips, ci=None)
plt.show()
```

```
sns.scatterplot(x="total_bill", y='tip', data=tips)   #the relationship between x and y
```

```
sns.scatterplot(x="total_bill", y="tip", hue="smoker", hue_order=['Yes','No'],data=tips)
plt.show()
```

```
plt.figure(figsize=(20,10))
sns.scatterplot(x="total_bill", y="tip", 
                hue="smoker", 
                hue_order=['Yes','No'],
                size='day',
                style='sex',
                data=tips)
plt.show()
```

```

sns.relplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```
sns.relplot(x='total_bill', y='tip', data=tips,
            kind='scatter', col="smoker")
plt.show()
```

```
sns.relplot(x='total_bill', y='tip', data=tips,
            kind='scatter', col="smoker", row='time')
plt.show()
```

## Coronavirus Sample



```
path = "https://raw.githubusercontent.com/RamiKrispin/coronavirus/master/csv/coronavirus.csv"
df = pd.read_csv(path)
df.head()
```

```
df.shape
```

```
sns.countplot(x=df.type)
plt.show()
```

```
sns.set(rc={'figure.figsize': (40,20)})
sns.barplot(x="country", y="cases",ci=None,data=df)
plt.xticks(rotation=90)
plt.show()
```

```
sns.set(rc={'figure.figsize': (40,20)})
sns.barplot(x="date", y="cases",ci=None,data=df)
plt.xticks(rotation=90)
plt.show()
```

```
df1 = df[(df['country']=="US" )|(df['country']=="Russia")|(df['country']=="Brazil")]
df1.shape
```

```
sns.lineplot(x="date", y="cases",hue="country", ci=None, data=df1)
plt.show()
```

```
sns.barplot(x="country", y="cases", hue="type",ci=None, data=df1)
plt.show()
```

```
sns.relplot(x="date", y="cases", row="country",
            height=5, aspect=5,
            ci=None, 
            kind="line", data=df1)
plt.xticks(rotation=90)
plt.show()
```
